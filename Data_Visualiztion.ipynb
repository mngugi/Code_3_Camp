```python
import pandas as pd

# Assuming you have a DataFrame called 'df'

# Calculate Q1, Q3, and IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Detect outliers
outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))

# Filter rows with outliers
outliers_df = df[outliers.any(axis=1)]

# Print or use outliers_df as needed
print("Rows with outliers:")
print(outliers_df)
```

```python
threshold = 0.5
drop_cols = df.columns[df.isnull().sum().mean() > threshold]
print(drop_cols)
```

```python
threshold = 0.5
drop_cols = df.columns[df.isnull().mean() > threshold]
df.drop(columns=drop_cols, inplace=True)

print("Dropped Columns:", drop_cols)
print("Original DataFrame with Columns Dropped:")
print(df)
```
