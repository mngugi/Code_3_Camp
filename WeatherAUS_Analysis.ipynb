```python
weather_Data['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
weather_Data['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)

fig = plt.figure(figsize = (8,5))
weather_Data.RainTomorrow.value_counts(normalize = True).plot(kind='bar', color= ['lightgray','dogerb'], alpha = 0.9, rot=0)
plt.title('RainTomorrow Indicator No(0) and Yes(1) in the Imbalanced Dataset')
plt.show()
```
