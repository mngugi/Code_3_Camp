```python
weather_Data['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
weather_Data['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)

fig = plt.figure(figsize = (8,5))
weather_Data.RainTomorrow.value_counts(normalize = True).plot(kind='bar', color= ['lightgray','skyblue'], alpha = 0.9, rot=0)
plt.title('RainTomorrow Indicator No(0) and Yes(1) in the Imbalanced Dataset')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame
# Replace 'df' with the actual variable name if it's different

# Selecting a subset of columns for illustration
selected_columns = ['MinTemp', 'MaxTemp', 'Rainfall', 'WindSpeed9am', 'Humidity9am', 'Temp9am']

# Creating subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))
fig.suptitle('Weather Data Subplots', fontsize=16)

# Plotting each selected column on a separate subplot
for i, col in enumerate(selected_columns):
    row, col_num = divmod(i, 3)
    axes[row, col_num].hist(weather_Data[col].dropna(), bins=30, color='skyblue', edgecolor='black')
    axes[row, col_num].set_title(col)
    axes[row, col_num].set_xlabel('Value')
    axes[row, col_num].set_ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Display the plots
plt.show()
```
