```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

dt = np.random.rand(10, 12)
sns.heatmap(dt, cmap='Oranges')

plt.show()
```

### Explanation
1. Import Required Libraries:

- seaborn for heatmap visualization.
- matplotlib.pyplot for displaying the heatmap.
- numpy for generating a random dataset.

2. Generate Random Data:

- np.random.rand(10, 12) creates a 10Ã—12 matrix with random values between 0 and 1.

3. Create a Heatmap:

- sns.heatmap(dt, cmap='Oranges') generates the heatmap.
- The cmap='Oranges' argument applies an orange color gradient.

4. Display the Heatmap:

- plt.show() renders the visualization.
