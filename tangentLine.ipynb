### Tangent Lines: The Slope of a Graph at a Point
- A tangent line to a curve at a given point is a straight line that "just touches" the curve at that point without crossing it (locally). - The slope of the tangent line at that point represents the instantaneous rate of change of the curve.

Key Concepts
1. Slope of a Curve:

For a straight line, the slope is constant and calculated as:

2. The slope of a line (\(m\)) is calculated as:

\[
m = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1}
\]

 
For a curve, the slope changes at every point. The slope at a specific point is defined as the slope of the tangent line to the curve at that point.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2  # Example: y = x^2

def f_prime(x):
    return 2*x  # Derivative: y' = 2x

# Point at which to calculate the tangent
x_point = 2
y_point = f(x_point)
slope = f_prime(x_point)

# Equation of the tangent line: y = m(x - x1) + y1
def tangent_line(x):
    return slope * (x - x_point) + y_point

# Generate x values for the curve and tangent line
x = np.linspace(0, 4, 100)  # Adjust range as needed
curve_y = f(x)
tangent_y = tangent_line(x)

# Plot the curve
plt.plot(x, curve_y, label="y = x^2", color="blue")

# Plot the tangent line
plt.plot(x, tangent_y, label=f"Tangent at x = {x_point}", color="red", linestyle="--")

# Mark the point of tangency
plt.scatter([x_point], [y_point], color="black", label=f"Point ({x_point}, {y_point})")

# Add labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.title("Tangent Line to y = x^2")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```
